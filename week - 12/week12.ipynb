### Import Libraries

```python
import cv2 
import os
import matplotlib.pyplot as plt
```

### 1. Lakukan Face Detection untuk image object lain yang tersedia pada (/images/facedet).

```python
cascade_wajah = cv2.CascadeClassifier('./haarcascades/haarcascade_frontalface_alt.xml')

# Directory containing images
image_directory = './Dataset/'

# List of images to process
images_to_process = [
    'jokowi.jpg',
    'kartini.jpg',
    'kucing.jpg',
    'mask.png',
    'mjordan.jpg',
    'solvayconf.jpg'
    # Add more image filenames here as needed
]

# Set up Matplotlib figure for subplots
num_images = len(images_to_process)
fig, axes = plt.subplots(nrows=num_images, ncols=2, figsize=(10, 5 * num_images))

# Loop through selected images
for i, filename in enumerate(images_to_process):
    img_path = os.path.join(image_directory, filename)
    
    # Check if the image exists in the directory
    if os.path.isfile(img_path):
        original_image = cv2.imread(img_path)
        original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib
        
        # Convert the image to grayscale (required for face detection)
        gray = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
        
        # Perform face detection
        faces = cascade_wajah.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
        
        # Draw rectangles around the detected faces on the original image
        for (x, y, w, h) in faces:
            cv2.rectangle(original_image, (x, y), (x + w, y + h), (0, 255, 0), 2)
        
        # Display the original image and the image with detected faces in subplots
        axes[i, 0].imshow(original_image_rgb)
        axes[i, 0].set_title('Original')
        axes[i, 0].axis('off')  # Hide axis for original image
        
        # Display the image with detected faces on the right
        detected_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib
        axes[i, 1].imshow(detected_image_rgb)
        axes[i, 1].set_title('Detected Faces')
        axes[i, 1].axis('off')  # Hide axis for detected faces
    else:
        print(f"File {filename} not found.")

# Adjust layout and show the subplots
plt.tight_layout()
plt.show()
```
