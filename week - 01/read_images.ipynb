# Langkah 1

### Import Libraries

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

# Langkah 2

### Reading Images from URLs

```python
urls = [
  "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
  "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
  "https://placekitten.com/800/571"
]
```

# Langkah 3

```python
for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat([image, image_2])
  cv2_imshow(final_frame)
  print('\n')
```

# Langkah 3

```python
height = image_2.shape[0]
width = image_2.shape[1]
print("Height x Width :", height, 'x', width)
```

# Langkah 4

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(width):
  image_3[int((height)/2), y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
print('\n')
```

# Langkah 5

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for x in range(height):
  image_3[x, int((width)/2)] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
print('\n')
```

# Langkah 6

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(height):
  image_3[y, int(y*(width/height))] = [255,255,255]
  image_3[y, width-1-int(y*(width/height))] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
print('\n')
```

# Langkah 7

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(100, 200):
  image_3[int((height)/2), y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
print('\n')
```

# Langkah 8

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(100, 200):
  for x in range(100, 200):
    image_3[x, y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
print('\n')
```
