### Import Libraries

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
```

### Load Image

```python
img = cv.imread('./KTP.jpeg')
img_plt = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv2_imshow(img)
plt.imshow(img_plt)
```

#### Censoring the section using rectangle

```python
cv.rectangle(img, pt1=(190, 325), pt2=(485, 345), color=(0,0,0), thickness=cv.FILLED) # Menutup bagian pekerjaan
cv.rectangle(img, pt1=(190, 350), pt2=(300, 367), color=(255,255,255), thickness=cv.FILLED) # Menutup bagian kewarganegaraan
cv2_imshow(img)
```
