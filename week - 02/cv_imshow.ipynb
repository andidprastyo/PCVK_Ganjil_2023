```python
import cv2 as cv 
from IPython.display import Image, display
```

### using display (on-cell)

```python
img = cv.imread('./KTP.jpeg')
# Display the image using cv2_imshow
cv.imwrite('image.jpeg', img)

# Wait for a key press and then close the window
display(Image('image.jpeg'))
```

### Using Imshow (pop=up)

```python
gambar = cv.imread('./KTP.jpeg')
cv.imshow('KTP', gambar)
cv.waitKey(0)
cv.destroyAllWindows()
```
